# Creating data

There are two core objects in pandas: the DataFrame and the Series.

# DataFrame

A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. 

Each entry corresponds to a row (or record) and a column.

```python
# Create a simple dataframe
data = {
    'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'year': [2012, 2012, 2013, 2014, 2014],
    'reports': [4, 24, 31, 2, 3]
}
df = pd.DataFrame(data, index = ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'])

print(df)

```

```python
data_1 = pd.DataFrame({'Name': ['Benson'], 'Age': [23], 'Marital Status': ['Single']}, index=['Benson'])

print(data_1)
```

# Series

If a DataFrame is a table, a Series is a list.

```python
# Create a series
s = pd.Series([3, 4, 6, 8, 10])

print(s)
```
